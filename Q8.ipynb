```python
from bs4 import BeautifulSoup
import requests
import re


paper_url = 'https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles'
resp = requests.get(paper_url)
resp
```

```python
resp.content
```

```python
soup1 = BeautifulSoup(resp.content)
soup1
```

```python
paper = soup1.find_all('h4', class_='italic')
paper
```

```python
paper_name = []
for i in paper:
    paper_name.append(i.text)
paper_name[:5]
```

```python
paper_name = []
for i in paper:
    paper_name.append(i.text.replace('\n',''))
paper_name[:5]
```

```python
author = soup1.find_all('p', class_ = 'sans bold')
author
```

```python
author_name = []
for i in author:
    author_name.append(i.text)
author_name[:5]
```

```python
author_name = []
for i in author:
    author_name.append(i.text.replace('\n',''))
author_name[:5]
```

```python
public_date = soup1.find_all('date', class_ = ('\((.*?)\)'))
public_date
```

```python
public_date = []
for i in public_date:
    public_date.append(i.text)
public_date[:5]
```

```python
public_date = []
for i in public_date:
    public_date.append(i.text.replace('\n',''))
public_date[:5]
```

```python
paper_link = soup1.find_all('a', class_ = 'paper_link')
paper_link
```

```python
paper_link = []
for i in paper_link:
    paper_link.append(i.text)
paper_link[:5]
```

```python
paper_link = []
for i in paper_link:
    paper_link.append(i.text.replace('\n',''))
paper_link[:5]
```
