1. Write a python program to display all the header tags from‘en.wikipedia.org/wiki/Main_Page’.


```python
#importing the important libraries for Web Scraping.

import  requests
from  bs4 import BeautifulSoup
```

```python
#Getting the request to the webpage to get the source code of the page
get_url = requests.get('https://en.wikipedia.org/wiki/Main_Page')
get_url
```

```python
# The web page source code downloaded above look like :
get_url.content
```

```python
# Parse the source code by using Beautiful Soup and the parser here ised is html.parser as the source code is in HTML.
soup_webpage = BeautifulSoup(get_url.content)
soup_webpage
```
